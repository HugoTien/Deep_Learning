```python
import matplotlib.pyplot as plt # 匯入plot_image並建立plot_image函數，讓數字影像能顯示出來
def plot_image(image):          # 設定傳入image為參數
    fig = plt.gcf()             # 設定fig
    fig.set_size_inches(2,2)    # 設定顯示圖案大小(inches)
    plt.imshow(image, cmap = 'binary') 
    # 設定使用plt.imshow圖型，
    # 參數image是28*28的圖型
    # cmap參數設定為binary，以黑白灰階表示
    plt.show() # 開始繪圖
```

```python
plot_image(x_train_image[0]) # 繪製 x_train_image的第一張圖，以及其對應的label
y_train_label[0]             # 繪製 x_train_image的第一張圖，以及其對應的label
```

```python
import matplotlib.pyplot as plt # 匯入pyplot模組，並命名plot
def plot_images_labels_prediction(images, labels, prediction,idx, num=10):
    fig = plt.gcf() # ????
    fig.set_size_inches(12,14)
    if num>25: num=25   # 如果顯示超過25筆，設定為25(避免錯誤)
    for i in range(0,25):
        ax = plt.subplot(5,5,1+i) # 建立subgraph子圖型為5行5列 (1+i?)
        ax.imshow(images[idx],cmap='binary') # 繪製subgraph子圖型
        title = "label=" +str(labels[idx])  # 設定子圖型title，並顯示標籤欄位
        if len(prediction) > 0: #  如果有傳入預測結果
            title+=",predict="+str(prediction[idx]) # 標題title加入預測結果
        ax.set_title(title,fontsize=10) # 設定子圖型的標題title與大小
        ax.set_xticks([]); ax.set_yticks([]) # 不顯示xy刻度
        idx+=1 #讀取下一筆
    plt.show() # 開始繪圖
```

```python
plot_images_labels_prediction(x_train_image,y_train_label,[],0,10)
```

```python
x_Train = x_train_image.reshape(60000, 784).astype('float32')
x_Test = x_test_image.reshape(10000, 784).astype('float32')
# 將60000張Train圖片以及1000張Test的圖片，從二維的28*28影像。轉換為一維的向量，再astype為float，共784個float數字
```

```python
x_train_image[0] # 查看train images的第一筆內容，數字為0~255(why)
```

```python
x_Train_normalize = x_Train/ 255
x_Test_normalize = x_Test/ 255  # 為了提高後續訓練的效率與準確率進行標準化，除255是最簡單的方法
```

```python
y_train_label[:5] # 訓練資料的前五筆data
```

```python
y_TrainOneHot = np_utils.to_categorical(y_train_label)
y_TestOneHot = np_utils.to_categorical(y_test_label)  # 將訓練資料與測試資料的標籤欄位，執行one-hot-encoding轉換
```

```python
y_TrainOneHot[:5] #看前五筆
```

# 建立模型

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential() # 建立線性堆疊模型
```

```python
model.add(Dense(units=256, # 定義隱藏層神經元數量
                input_dim=784, # 設定輸入層784(28*28)
                kernel_initializer= 'normal', # 使用常態分布亂數，初始化weigght, bias(?)
                activation = 'relu')) # 定義機活參數為relu (??)
```

```python
model.add(Dense(units=10, # 定輸出層神經元數量 
                kernel_initializer = 'normal',
                activation='softmax')) # 定義激活參數為softmax (?)
```

```python
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',metrics=['accuracy'])
# loss為損失函數，通常深度學習使用cross_entropy(訓練效果較好)
# 設定訓練時的最優化方法，adam可以讓訓練更快收斂，並提高準確度
# metrics: 設定評估模型方式是accuracy
```

```python
train_history = model.fit(x=x_Train_normalize, # features數字影像的特徵值
                          y=y_TrainOneHot,    # label數字影像的真實值
                          validation_split = 0.2, # 20% 驗證
                          epochs = 10, # 訓練週期執行十次
                          batch_size = 200, # 每次訓練200筆
                          verbose =2)  # 顯示訓練過程
```

```python
import matplotlib.pyplot as plt
def show_train_history(train_history, train, validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train', 'validation'], loc = 'upper left')
    plt.show # 畫圖(待研究)
```

```python
scores = model.evaluate(x_Test_normalize, y_TestOneHot) # 評估模型準確率
```

```python
prediction = model.predict_classes(x_Test) # 執行預測準確率
```

```python
plot_images_labels_prediction(x_test_image, y_test_label,
                             prediction, idx=340)
```
